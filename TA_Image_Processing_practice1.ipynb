```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# imread command

```python
image = cv2.imread('/content/drive/MyDrive/Jujutsu Kaisen Suguru.jfif')
```

# imshow command

```python
cv2_imshow(image)
```

# resize command

```python
resized_image = cv2.resize(image,(250,250))
cv2_imshow(resized_image)
```

# gray color command

```python
gray_color_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray_color_image)
```

# histogram image command




```python
his_image = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.hist(image.ravel(), 256, [0, 256])
plt.title('Histogram for image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

# histogram gray_image command

```python
his_image = cv2.calcHist([gray_color_image], [0], None, [256], [0, 256])

plt.hist(gray_color_image.ravel(), 256, [0, 256])
plt.title('Histogram for gray_color_image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

# save the gray_color_image on google drive

```python
gray_color_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gray_color_image = cv2.resize(gray_color_image , (250,250))
cv2_imshow(gray_color_image)

cv2.imwrite('/content/drive/MyDrive/grayscale_image.jpg' , gray_color_image)
```
